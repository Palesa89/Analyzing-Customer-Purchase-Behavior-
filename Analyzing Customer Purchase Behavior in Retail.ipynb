```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'])

# Find the range of dates
start_date = df['transaction_date'].min()
end_date = df['transaction_date'].max()
print(f"Date Range: {start_date} to {end_date}")
```

```python
# Ensure the 'purchase_date' column is in datetime format
df['transaction_date'] = pd.to_datetime(df['transaction_date'])

# Extract month, day of the week, and year
df['month'] = df['transaction_date'].dt.month
df['day_of_week'] = df['transaction_date'].dt.day_name()
df['year'] = df['transaction_date'].dt.year

# Aggregate sales data by month
monthly_sales = df.groupby(['year', 'month'])['total_sales'].sum().reset_index()

plt.figure(figsize=(12, 6))
for year in monthly_sales['year'].unique():
    monthly_data = monthly_sales[monthly_sales['year'] == year]
    plt.plot(monthly_data['month'], monthly_data['total_sales'], marker='o', label=str(year))

plt.title('Monthly Sales Over Years')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title='Year')
plt.grid()
plt.show();
```

```python
# Group by product_id to calculate total sales, total quantity sold, and average rating
product_performance = df.groupby('product_id').agg({
    'quantity': 'sum',  # Total quantity sold
    'total_sales': 'sum',  # Total sales revenue
    'product_rating': 'mean'  # Average product rating
}).reset_index()

# Sort products by total sales to find top-performing products
top_products_by_sales = product_performance.sort_values(by='total_sales', ascending=False)

# Sort products by total quantity sold to find top-performing products in terms of units sold
top_products_by_quantity = product_performance.sort_values(by='quantity', ascending=False)

# Sort products by average rating to find products with the best customer reviews
top_products_by_rating = product_performance.sort_values(by='product_rating', ascending=False)
```

```python
# Top 10% performers by sales
top_10_percent_sales = product_performance['total_sales'].quantile(0.90)
top_performers_sales = product_performance[product_performance['total_sales'] >= top_10_percent_sales]

# Top 10% performers by quantity sold
top_10_percent_quantity = product_performance['quantity'].quantile(0.90)
top_performers_quantity = product_performance[product_performance['quantity'] >= top_10_percent_quantity]

# Top 10% performers by average rating
top_10_percent_rating = product_performance['product_rating'].quantile(0.90)
top_performers_rating = product_performance[product_performance['product_rating'] >= top_10_percent_rating]

print("Top Performing Products by Sales:")
print(top_performers_sales)

print("Top Performing Products by Quantity Sold:")
print(top_performers_quantity)

print("Top Performing Products by Average Rating:")
print(top_performers_rating)
```

```python
# Bottom 10% performers by sales
bottom_10_percent_sales = product_performance['total_sales'].quantile(0.10)
underperformers_sales = product_performance[product_performance['total_sales'] <= bottom_10_percent_sales]

# Bottom 10% performers by quantity sold
bottom_10_percent_quantity = product_performance['quantity'].quantile(0.10)
underperformers_quantity = product_performance[product_performance['quantity'] <= bottom_10_percent_quantity]

# Bottom 10% performers by rating
bottom_10_percent_rating = product_performance['product_rating'].quantile(0.10)
underperformers_rating = product_performance[product_performance['product_rating'] <= bottom_10_percent_rating]

print("Underperforming Products by Sales:")
print(underperformers_sales)

print("Underperforming Products by Quantity Sold:")
print(underperformers_quantity)

print("Underperforming Products by Rating:")
print(underperformers_rating)
```

```python
# Calculate purchase frequency per customer (if not already available in your data)
customer_purchase_frequency = df.groupby('customer_zip_code').agg({
    'total_transactions': 'sum',  
    'avg_transaction_value': 'mean'  
}).reset_index()

print(customer_purchase_frequency.head())
```

```python
# Calculate repeat purchases and CLV
df['total_spent'] = df['avg_transaction_value'] * df['total_transactions']  # Total spent by each customer

# Calculate Customer Lifetime Value (CLV) for each customer
customer_lifetime_value = df.groupby('customer_zip_code').agg({
    'total_spent': 'sum',  
    'total_transactions': 'sum'  
}).reset_index()

# Calculate CLV as total revenue (sum of all transactions) / total transactions (or with average values)
customer_lifetime_value['CLV'] = customer_lifetime_value['total_spent'] / customer_lifetime_value['total_transactions']

print(customer_lifetime_value.head())
```

```python
# Analyze churn by grouping customers who churned vs. those who did not
churn_analysis = df.groupby('churned').agg({
    'days_since_last_purchase': 'mean',  
    'avg_discount_used': 'mean',  
    'loyalty_program': lambda x: x.value_counts(normalize=True).get('Yes', 0)  
}).reset_index()

print(churn_analysis)
```

```python
# Combine customer behavior, CLV, and churn indicators
customer_analysis = df.groupby('customer_zip_code').agg({
    'total_transactions': 'sum',  
    'avg_transaction_value': 'mean',  
    'total_spent': 'sum',  
    'churned': lambda x: x.mode()[0],  
    'days_since_last_purchase': 'mean', 
    'avg_discount_used': 'mean',  
}).reset_index()

customer_analysis['CLV'] = customer_analysis['total_spent'] / customer_analysis['total_transactions']

print(customer_analysis.head())
```

```python
# Set thresholds for identifying potential counterfeit products
low_rating_threshold = 2.5  
high_return_rate_threshold = 0.2  
low_price_threshold = df['unit_price'].quantile(0.1)  
# Identify potentially counterfeit products
df['potential_counterfeit'] = ((df['product_rating'] < low_rating_threshold) & 
                               (df['product_return_rate'] > high_return_rate_threshold) & 
                               (df['unit_price'] < low_price_threshold)).astype(int)

# Display the number of potentially counterfeit products
print(f"Total Potentially Counterfeit Products: {df['potential_counterfeit'].sum()}")
```

```python
# Group by counterfeit vs authentic status and calculate total sales
sales_comparison = df.groupby('potential_counterfeit').agg({
    'total_sales': 'sum',  
    'quantity': 'sum'  
}).reset_index()

sales_comparison['type'] = sales_comparison['potential_counterfeit'].replace({0: 'Authentic', 1: 'Potentially Counterfeit'})

print(sales_comparison)
```

```python
# Filter for counterfeit products
counterfeit_products = df[df['potential_counterfeit'] == 1]

# Find the most common counterfeit product categories
common_counterfeit_categories = counterfeit_products.groupby('product_category').agg({
    'product_id': 'count',  
    'total_sales': 'sum'  
}).reset_index()

# Sort by the most common counterfeit categories
common_counterfeit_categories = common_counterfeit_categories.sort_values(by='product_id', ascending=False)

print("Most Common Counterfeit Product Categories:")
print(common_counterfeit_categories.head(10))
```

```python
# Bar plot of sales comparison between counterfeit and authentic products
plt.figure(figsize=(8, 5))
plt.bar(sales_comparison['type'], sales_comparison['total_sales'], color=['green', 'orange'])
plt.xlabel('Product Type')
plt.ylabel('Total Sales')
plt.title('Sales Comparison: Counterfeit vs Authentic Products')
plt.show()
```

```python
# Aggregate sales based on holiday seasons
sales_by_holiday = df.groupby('holiday_season').agg({
    'total_sales': 'sum',  # Total sales during holiday vs. non-holiday
    'total_transactions': 'count'  # Total number of transactions
}).reset_index()

print("Sales by Holiday Season:")
print(sales_by_holiday)
```

```python
# Aggregate sales based on seasons
sales_by_season = df.groupby('season').agg({
    'total_sales': 'sum',
    'total_transactions': 'count'
}).reset_index()

print("Sales by Season:")
print(sales_by_season)
```

```python
# Aggregate sales by day of the week
sales_by_day = df.groupby('day_of_week').agg({
    'total_sales': 'sum',
    'total_transactions': 'count'
}).reset_index()

# Ensure the days are in order
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sales_by_day['day_of_week'] = pd.Categorical(sales_by_day['day_of_week'], categories=days_order, ordered=True)
sales_by_day = sales_by_day.sort_values('day_of_week')

print("Sales by Day of the Week:")
print(sales_by_day)
```
